```python
my_list = [10, 3.5, "hello", True]

print("List:", my_list)
print("Type of list:", type(my_list))

for element in my_list:
    print(element, ":", type(element))
```

```python
a = [1, 2, 3]
b = [4, 5, 6]


print("Concatenation:", a + b)

# Repetition
print("Repetition:", a * 3)

# Membership check
print("Is 2 in list a?", 2 in a)

# Length
print("Length of list a:", len(a))
```

```python
nums = [10, 5, 20, 8, 15]

print("Maximum:", max(nums))
print("Minimum:", min(nums))
print("Sum:", sum(nums))


print("Ascending:", sorted(nums))
print("Descending:", sorted(nums, reverse=True))


print("Index of 8:", nums.index(8))
```

```python
my_list = [10, 20, 30]

my_list.append(40)        # Add element
my_list.insert(1, 15)     # Insert at index 1
my_list.remove(20)        # Remove element
my_list.pop()             # Pop last element
my_list.reverse()         # Reverse list

print("Final list:", my_list)
```

```python

nums = list(map(int, input("Enter numbers separated by space: ").split()))

evens = [n for n in nums if n % 2 == 0]
print("Even numbers:", evens)


odds = [n for n in nums if n % 2 != 0]
print("Odd numbers:", odds)


print("Largest:", max(nums))
print("Smallest:", min(nums))


unique_nums = list(set(nums))
print("Without duplicates:", unique_nums)
```
