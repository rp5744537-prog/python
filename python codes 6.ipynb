```python
def display_menu():
    print("\n--- Train Seat Reservation System ---")
    print("1. Book Seat")
    print("2. Cancel Seat")
    print("3. Show Seats")
    print("4. Booking Percentage")
    print("5. Exit")

def book_seat(seats):
    seat_no = int(input("Enter seat number to book (1-20): "))
    if 1 <= seat_no <= 20:
        if seats[seat_no - 1] == 0:
            seats[seat_no - 1] = 1
            print(f"Seat {seat_no} booked successfully.")
        else:
            print(f"Seat {seat_no} is already booked.")
    else:
        print("Invalid seat number.")

def cancel_seat(seats):
    seat_no = int(input("Enter seat number to cancel (1-20): "))
    if 1 <= seat_no <= 20:
        if seats[seat_no - 1] == 1:
            seats[seat_no - 1] = 0
            print(f"Seat {seat_no} cancelled successfully.")
        else:
            print(f"Seat {seat_no} is not booked.")
    else:
        print("Invalid seat number.")

def show_seats(seats):
    print("\nSeat Status (0 = Available, 1 = Booked):")
    for i, seat in enumerate(seats, start=1):
        print(f"Seat {i}: {seat}")

def booking_percentage(seats):
    booked = seats.count(1)
    percentage = (booked / len(seats)) * 100
    print(f"\nBooking Percentage: {percentage:.2f}%")

def main():
    seats = [0] * 20  # Initialize 20 seats as available
    while True:
        display_menu()
        choice = int(input("Enter your choice: "))
        
        if choice == 1:
            book_seat(seats)
        elif choice == 2:
            cancel_seat(seats)
        elif choice == 3:
            show_seats(seats)
        elif choice == 4:
            booking_percentage(seats)
        elif choice == 5:
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
def analyze_playlist(videos):
    total_time = sum(videos)  # Total watch time
    longest = max(videos)     # Longest video
    average = total_time / len(videos)  # Average duration
    short_videos = [v for v in videos if v < 5]  # Videos shorter than 5 minutes

    print("\n--- YouTube Playlist Analysis ---")
    print(f"Total watch time: {total_time} minutes")
    print(f"Longest video: {longest} minutes")
    print(f"Average video duration: {average:.2f} minutes")
    print(f"Videos shorter than 5 minutes: {short_videos}")

# Sample Input
playlist = [12, 4, 8, 15, 3, 20]
analyze_playlist(playlist)
```

```python
def hashtag_tracker(hashtags):
    # Frequency dictionary
    freq = {}
    for tag in hashtags:
        freq[tag] = freq.get(tag, 0) + 1

    # Most repeated hashtag
    most_repeated = max(freq, key=freq.get)

    # Unique hashtags
    unique_tags = list(set(hashtags))

    # Remove duplicates (preserve order)
    seen = set()
    no_duplicates = []
    for tag in hashtags:
        if tag not in seen:
            no_duplicates.append(tag)
            seen.add(tag)

    print("\n--- Social Media Hashtag Tracker ---")
    print(f"Most repeated hashtag: {most_repeated}")
    print(f"Unique hashtags: {unique_tags}")
    print("Frequency of each hashtag:")
    for tag, count in freq.items():
        print(f"{tag}: {count}")
    print(f"List without duplicates: {no_duplicates}")

# Sample Input
hashtags = ["#fun", "#python", "#fun", "#code", "#python"]
hashtag_tracker(hashtags)
```

```python
def expense_splitter(expenses):
    total = sum(expenses)  # Total expense
    average_share = total / len(expenses)  # Equal share
    highest_paid = max(expenses)  # Highest contribution
    highest_index = expenses.index(highest_paid) + 1  # Friend number (1-based)

    print("\n--- Expense Splitter ---")
    print(f"Total expense: {total}")
    print(f"Average share: {average_share:.2f}")
    print(f"Person who paid highest: Friend {highest_index} (Paid {highest_paid})")
    print("\nSettlement among friends:")
    for i, amount in enumerate(expenses, start=1):
        balance = average_share - amount
        if balance > 0:
            print(f"Friend {i} should pay {balance:.2f}")
        elif balance < 0:
            print(f"Friend {i} should receive {abs(balance):.2f}")
        else:
            print(f"Friend {i} is settled.")

# Sample Input
expenses = [1200, 850, 1500, 950]
expense_splitter(expenses)
```

```python
# Pattern Compression

def pattern_compression(lst):
    compressed = []
    i = 0
    while i < len(lst):
        count = 1
        # Count consecutive duplicates
        while i + 1 < len(lst) and lst[i] == lst[i + 1]:
            count += 1
            i += 1
        compressed.append((lst[i], count))
        i += 1
    return compressed

# Sample Input
data = [1, 1, 1, 2, 2, 3, 3, 3]

# Run compression
result = pattern_compression(data)

print("\n--- Pattern Compression ---")
print("Input:", data)
print("Output:", result)
```

```python
# Mini Shopping Cart System

def display_menu():
    print("\n--- Mini Shopping Cart ---")
    print("1. Add Item")
    print("2. Remove Item")
    print("3. Search Item")
    print("4. Calculate Total Bill")
    print("5. Show Duplicate Items")
    print("6. Exit")

def add_item(cart, prices):
    item = input("Enter item name: ")
    price = float(input(f"Enter price for {item}: "))
    cart.append(item)
    prices.append(price)
    print(f"{item} added successfully.")

def remove_item(cart, prices):
    item = input("Enter item name to remove: ")
    if item in cart:
        index = cart.index(item)
        cart.pop(index)
        prices.pop(index)
        print(f"{item} removed successfully.")
    else:
        print(f"{item} not found in cart.")

def search_item(cart):
    item = input("Enter item name to search: ")
    if item in cart:
        print(f"{item} found in cart at position {cart.index(item)+1}.")
    else:
        print(f"{item} not found in cart.")

def calculate_bill(prices):
    total = sum(prices)
    print(f"Total Bill: {total:.2f}")

def show_duplicates(cart):
    duplicates = []
    for item in cart:
        if cart.count(item) > 1 and item not in duplicates:
            duplicates.append(item)
    if duplicates:
        print("Duplicate items in cart:", duplicates)
    else:
        print("No duplicate items found.")

def main():
    cart = []
    prices = []
    while True:
        display_menu()
        choice = int(input("Enter your choice: "))
        
        if choice == 1:
            add_item(cart, prices)
        elif choice == 2:
            remove_item(cart, prices)
        elif choice == 3:
            search_item(cart)
        elif choice == 4:
            calculate_bill(prices)
        elif choice == 5:
            show_duplicates(cart)
        elif choice == 6:
            print("Exiting Shopping Cart. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
