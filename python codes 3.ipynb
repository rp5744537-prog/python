```python
def two_sum_optimized(nums, target):
    
    num_map = {} 
   
    for i, num in enumerate(nums):

        complement = target - num
        
        if complement in num_map:
            
            return [num_map[complement], i]
        
        num_map[num] = i
    
    return []  
 
nums = [2, 7, 11, 15]
target = 9
print(two_sum_optimized(nums, target))
```

```python
def merge(intervals):
    intervals.sort(key=lambda i: i[0])
    output = [intervals[0]]
    for start, end in intervals[1:]:
        lastEnd = output[-1][1]
        if start <= lastEnd:
            output[-1][1] = max(lastEnd, end)
        else:
            output.append([start, end])
    return output
intervals_input = [[1,3],[2,6],[8,10],[15,18]]
print("3. Merged Intervals:", merge(intervals_input))
```

```python
def productExceptSelf(nums):
    res = [1] * len(nums)
    prefix = 1
    for i in range(len(nums)):
        res[i] = prefix
        prefix *= nums[i]
    postfix = 1
    for i in range(len(nums) - 1, -1, -1):
        res[i] *= postfix
        postfix *= nums[i]
    return res

# Test
nums_prod = [1, 2, 3, 4]
print("4. Product Except Self:", productExceptSelf(nums_prod))
```

```python
def isValid(s):
    Map = {")": "(", "]": "[", "}": "{"}
    stack = []
    for c in s:
        if c in Map:
            if stack and stack[-1] == Map[c]:
                stack.pop()
            else:
                return False
        else:
            stack.append(c)
    return True if not stack else False

# Test
test_str = "({[]})"
print("5. Is Valid Parentheses:", isValid(test_str))
```

```python
def search(nums, target):
    l, r = 0, len(nums) - 1
    while l <= r:
        mid = (l + r) // 2
        if target == nums[mid]: return mid
        if nums[l] <= nums[mid]:
            if target > nums[mid] or target < nums[l]: l = mid + 1
            else: r = mid - 1
        else:
            if target < nums[mid] or target > nums[r]: r = mid - 1
            else: l = mid + 1
    return -1


rotated_nums = [4,5,6,7,0,1,2]
print("7. Target Index in Rotated Array:", search(rotated_nums, 0))
```

```python
from collections import defaultdict

def groupAnagrams(strs):
    res = defaultdict(list)
    for s in strs:
        count = [0] * 26
        for c in s:
            count[ord(c) - ord("a")] += 1
        res[tuple(count)].append(s)
    return list(res.values())

words = ["eat","tea","tan","ate","nat","bat"]
print("8. Grouped Anagrams:", groupAnagrams(words))
```

```python
from collections import deque
def maxSlidingWindow(nums, k):
    output = []
    q = deque()
    l = r = 0
    while r < len(nums):
        while q and nums[q[-1]] < nums[r]:
            q.pop()
        q.append(r)
        if l > q[0]:
            q.popleft()
        if (r + 1) >= k:
            output.append(nums[q[0]])
            l += 1
        r += 1
    return output

print("9. Sliding Window Max:", maxSlidingWindow([1,3,-1,-3,5,3,6,7], 3))
```

```python
def findPlatform(arr, dep):
    arr.sort()
    dep.sort()
    plat_needed, result = 1, 1
    i, j = 1, 0
    while i < len(arr) and j < len(dep):
        if arr[i] <= dep[j]:
            plat_needed += 1
            i += 1
        else:
            plat_needed -= 1
            j += 1
        result = max(result, plat_needed)
    return result

print("10. Min Platforms:", findPlatform([900, 940, 950, 1100, 1500, 1800], [910, 1200, 1120, 1130, 1900, 2000]))
```

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

def hasCycle(head):
    slow, fast = head, head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast: return True
    return False

# Manual test setup
node1 = ListNode(1); node2 = ListNode(2)
node1.next = node2; node2.next = node1 # Cycle
print("11. Cycle Detected:", hasCycle(node1))
```

```python
from collections import Counter
def topKFrequent(nums, k):
    count = Counter(nums)
    return [item[0] for item in count.most_common(k)]

print("12. Top K Frequent:", topKFrequent([1,1,1,2,2,3], 2))
```

```python
def wordBreak(s, wordDict):
    dp = [False] * (len(s) + 1)
    dp[len(s)] = True
    for i in range(len(s) - 1, -1, -1):
        for w in wordDict:
            if (i + len(w)) <= len(s) and s[i : i + len(w)] == w:
                dp[i] = dp[i + len(w)]
            if dp[i]: break
    return dp[0]

print("13. Word Break:", wordBreak("leetcode", ["leet", "code"]))
```
